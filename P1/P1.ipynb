```python
import numpy as np
from pandas.io.parsers import read_csv

valores=read_csv("ex1data1.csv",header=None).to_numpy()
```

```python
def recta(w,b,x):
    return w*x+b
```

```python
def der_w(valores, w, b):
    m = len(valores)
    return (1/m)*np.sum((recta(w,b,valores[:,0])-valores[:,1])* valores[:,0])

def der_b(valores, w, b):
    m = len(valores)
    return (1/m)*np.sum(recta(w,b,valores[:,0])-valores[:,1])

def funcion_coste(valores, w, b):
    m = len(valores)
    return (1/(2*m))*np.sum((recta(w,b,valores[:,0])-valores[:,1])**2)
```

```python
def funcion_grad(valores, w, b, a, iter):
    for i in range(iter):  
        w = w - a*der_w(valores,w,b)
        b = b - a*der_b(valores,w,b)
    return w, b
```

```python
w = 0
b = 0
a = 0.01
iter = 1500

res = funcion_coste(valores, w, b)
res
```

```python
w, b = funcion_grad(valores, w, b, a, iter)
comprobacion = recta(w,b,7)
comprobacion
```
