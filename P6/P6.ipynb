```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import SparseCategoricalCrossentropy
```

```python
R_S = 42
Test_values = 1500
```

```python
auxX = np.load('data6/X.npy')
auxY = np.load('data6/y.npy')

X_ent, X_test, y_ent, y_test = train_test_split(auxX, auxY, test_size=0.3, random_state=R_S)
```

```python
model1 = Sequential([Dense(units = 25,activation = 'relu'), Dense(units = 15,activation = 'relu'), Dense(units = 10,activation = 'softmax')])
model1.compile(loss = SparseCategoricalCrossentropy(from_logits=True))
model1.fit(X_ent,y_ent,epochs = 100)

z1 = model1(X_ent)

f_x = tf.nn.softmax(z1)
```

```python
model2 = Sequential([Dense(units = 25,activation = 'relu'), Dense(units = 15,activation = 'relu'), Dense(units = 10,activation = 'softmax')])
model2.compile(optimizer= Adam(learning_rate=1e-3), loss = SparseCategoricalCrossentropy(from_logits=True))
model2.fit(X_ent,y_ent,epochs = 100)

z2 = model2(X_ent)

f_x = tf.nn.softmax(z2)
```

```python
model3 = Sequential([Dense(units = 25,activation = 'relu'), Dense(units = 15,activation = 'relu'), Dense(units = 10,activation = 'linear')])
model3.compile(loss = SparseCategoricalCrossentropy(from_logits=True))
model3.fit(X_ent,y_ent,epochs = 100)

z3 = model3(X_ent)

f_x = tf.nn.softmax(z3)
```

```python
model4 = Sequential([Dense(units = 25,activation = 'relu'), Dense(units = 15,activation = 'relu'), Dense(units = 10,activation = 'linear')])
model4.compile(optimizer= Adam(learning_rate = 1e-3), loss = SparseCategoricalCrossentropy(from_logits=True))
model4.fit(X_ent,y_ent,epochs = 100)

z4 = model4(X_ent)

f_x = tf.nn.softmax(z4)
```

```python
# Obtenemos las predicciones
predictions1 = model1.predict(X_test)
predictions2 = model2.predict(X_test)
predictions3 = model3.predict(X_test)
predictions4 = model4.predict(X_test)

results1 = np.argmax(predictions1, axis=-1, keepdims=True)
results2 = np.argmax(predictions2, axis=-1, keepdims=True)
results3 = np.argmax(predictions3, axis=-1, keepdims=True)
results4 = np.argmax(predictions4, axis=-1, keepdims=True)


#Vemos la precisión
print("Softmax y descenso de gradiente")
print("Precision: ", np.sum(results1 == y_test)/ len(y_test)*100, "%")
print("Erroneos: ", np.sum(results1 == y_test), " de ", len(y_test), "\n")

print("Softmax y adam")
print("Precision: ", np.sum(results2 == y_test)/ len(y_test)*100, "%")
print("Erroneos: ", np.sum(results2 == y_test), " de ", len(y_test),"\n")

print("Linear y descenso de gradiente")
print("Precision: ", np.sum(results3 == y_test)/ len(y_test)*100, "%")
print("Erroneos: ", np.sum(results3 == y_test), " de ", len(y_test),"\n")

print("Linear y adam")
print("Precision: ", np.sum(results4 == y_test)/ len(y_test)*100, "%")
print("Erroneos: ", np.sum(results4 == y_test), " de ", len(y_test),"\n")
```
